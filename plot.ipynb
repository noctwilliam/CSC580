```python
import matplotlib.pyplot as plt

# Hypothetical data
num_threads = [i for i in range (1,17)]
time_taken = [61,31,21,25,28,27,24,26,29,32,29,29,24,23,22,24]

# Plotting the graph
plt.plot(num_threads, time_taken, marker='o')

# Adding labels and title
plt.xlabel('Number of Threads')
plt.ylabel('Time Taken (ms)')
plt.title('Number of Threads vs. Time Taken (ms)')

# Display the graph
plt.show()
```
